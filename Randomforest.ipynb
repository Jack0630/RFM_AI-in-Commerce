```python
import numpy as np     # scientific calculation
import pandas as pd    # handle csv file
import seaborn as sns  # visualize 
import matplotlib.pyplot as plt # plotting
```

```python
df =pd.read_csv('RFM_clear.csv')
```

```python
df['RFM_Score'] = df.R.astype(str)+ df.F.astype(str) + df.M.astype(str)
df.drop(df.tail(1).index,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(labels=['InvoiceDate','R','F','M','RFM_Score'],axis=1).values
Y = df['RFM_Score'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=160)

print('train shape:', X_train.shape)
print('test shape:', X_test.shape)
```

```python
Y_test = Y_test.astype(int)
```

```python
from sklearn.ensemble import RandomForestRegressor

#10個決策樹
rf = RandomForestRegressor(n_estimators= 10, random_state=42)
rf.fit(X_train, Y_train)
```

```python
predictions = rf.predict(X_test)
predictions=np.around(predictions)
```

```python
predictions=predictions.astype(int)
```

```python
i=0
k=0

for n in range(len(predictions)):
    if predictions[n] == Y_test[n]:
        i+=1
    k+=1
    
rate= (i/k)*100
rate=str(round(rate,2))

print('precise rate : '+ rate + "%")
```
